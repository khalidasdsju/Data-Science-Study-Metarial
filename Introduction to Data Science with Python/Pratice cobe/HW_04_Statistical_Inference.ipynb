### Problem_01:[ One sample t-test]
Data: Systolic blood pressures of 14 patients are given below:
183, 152, 178, 157, 194, 163, 144, 114, 178, 152, 118, 158, 172, 138

Test, whether the population mean, is less than 165

**Hypothesis**
<br>H0: There is no significant mean difference in systolic blood pressure. i.e., μ = 165
<br>H1: The population mean is less than 165. i.e., μ < 165

**Level of Significance: 0.05**

**Test statistic**  
 

Where,
x̄ is sample mean
μ is the population mean
s is sample standard deviation
n is the number of observations;

```python
sys_bp=[183, 152, 178, 157, 194, 163, 144, 114, 178, 152, 118, 158, 172, 138]

mu=165

from scipy import stats

t_value,p_value=stats.ttest_1samp(sys_bp,mu)
```

```python
one_tailed_p_value=float("{:.6f}".format(p_value/2)) # Since alternative hypothesis is one tailed, We need to divide the p value by 2.

print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for one tailed test is %f'%one_tailed_p_value)

print('Test statistic is %.6f',t_value)

print('p-value for one tailed test is ',one_tailed_p_value)

alpha = 0.05

if one_tailed_p_value<=alpha:
  print('Conclusion','n','Since p value(=%f)'%p_value,'<','alpha(=%.2f)'%alpha,
        '''We reject the null hypothesis H0. So we conclude that there is no significant mean 
        difference in systolic blood pressure. i.e., μ = 165 at %.2f level of significance'''%alpha)

else:
  print('Conclusion','n','Sincep-value(=%f)'%one_tailed_p_value, '>', 'alpha(=%.2f)'%alpha,'We do not reject the null hypothesis H0.')
```

<br>H1: The effect of ammonium chloride and urea on grain yield of paddy is not equal i.e., μ1 ≠ μ2**

**Test statistic**

### Problem_05:
A bottle filling machine fills water into 5 liters (5000 cm3) bottles. The company wants to test the null
hypothesis that the average amount of water filled by the machine into the bottle is at least 5000 cm3.
A random sample of 60 bottles coming out of the machine was selected and the exact contents of the
selected bottles are recorded. The sample mean was 4,998.1 cm3. The population standard deviation is
known from the experience to be 1.30 cm3. Assume that the population is normally distributed with
the standard deviation of 1.30 cm3. Write code to test the hypothesis at a of 5%. Explain the results.

```python
import scipy.stats as stats

# Given data
sample_mean = 4998.1  # Sample mean
population_stddev = 1.30  # Population standard deviation
sample_size = 60  # Sample size
population_mean = 5000  # Hypothesized population mean

# Calculate the standard error of the sample mean
standard_error = population_stddev / (sample_size ** 0.5)

# Calculate the z-score
z_score = (sample_mean - population_mean) / standard_error

# Calculate the p-value
p_value = 1 - stats.norm.cdf(z_score)

# Set the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to make a decision
if p_value < alpha:
    print(f"Reject the null hypothesis. The average amount of water filled is less than 5000 cm^3 (p-value = {p_value})")
else:
    print(f"Fail to reject the null hypothesis. There is not enough evidence to conclude 
          that the average amount of water filled is less than 5000 cm^3 (p-value = {p_value})")
```

### Problem_06: [Find the Probability using z distribution]
As per a survey on use of pesticides among 1000 farmers in grape farming for around 10 acres of
grape farmland, it was found that the grape farmers spray 38 liters of pesticides in a week on an average
with the corresponding standard deviation of 5 liters. Assume that the pesticide spray per week
follows a normal distribution. Write code to answer the following questions:
<br>(a) What proportion of the farmers is spraying more than 50 liters of pesticide in a week?
<br>(b) What proportion of farmers is spraying less than 10 liters?
<br>(c) What proportion of farmers is spraying between 30 liters and 60 liters?

```python
import scipy.stats as stats

# Given data
mean = 38  # Mean pesticide spray per week
std_dev = 5  # Standard deviation
total_farmers = 1000  # Total number of farmers

# (a) What proportion of the farmers is spraying more than 50 liters of pesticide in a week?
x_a = 50  # Threshold for more than 50 liters
z_score_a = (x_a - mean) / std_dev  # Calculate the z-score
probability_a = 1 - stats.norm.cdf(z_score_a)  # Calculate the probability

# (b) What proportion of farmers is spraying less than 10 liters?
x_b = 10  # Threshold for less than 10 liters
z_score_b = (x_b - mean) / std_dev  # Calculate the z-score
probability_b = stats.norm.cdf(z_score_b)  # Calculate the probability

# (c) What proportion of farmers is spraying between 30 liters and 60 liters?
x_c_lower = 30  # Lower threshold
x_c_upper = 60  # Upper threshold
z_score_c_lower = (x_c_lower - mean) / std_dev  # Calculate the z-score for the lower threshold
z_score_c_upper = (x_c_upper - mean) / std_dev  # Calculate the z-score for the upper threshold
probability_c = stats.norm.cdf(z_score_c_upper) - stats.norm.cdf(z_score_c_lower)  # Calculate the probability

# Print the results
print(f"(a) Proportion of farmers spraying more than 50 liters: {probability_a:.4f}")
print(f"(b) Proportion of farmers spraying less than 10 liters: {probability_b:.4f}")
print(f"(c) Proportion of farmers spraying between 30 and 60 liters: {probability_c:.4f}")
```

### Problem_07. 
Suppose that the makers of ABC batteries want to demonstrate that their battery lasts an average of
at least 60 min longer than their competitor brand. Two independent random samples of 100 batteries
of each kind are selected from both the brands, and the batteries are used continuously. The
sample average life of ABC is found to be 450 min. The average life of competitor batteries is 368 min
with the sample standard deviation of 82 min and 78 min, respectively. Frame a hypothesis and write
the code to test ABC’s claim at 95% significance.


```python
import scipy.stats as stats

# Given data for ABC batteries
sample_mean_ABC = 450  # Sample mean life of ABC batteries
sample_size_ABC = 100  # Sample size for ABC batteries

# Given data for competitor brand batteries
sample_mean_Competitor = 368  # Sample mean life of competitor brand batteries
sample_size_Competitor = 100  # Sample size for competitor brand batteries
sample_stddev_Competitor = 78  # Sample standard deviation for competitor brand batteries

# Hypothesized difference in means (ABC - Competitor)
hypothesized_difference = 60

# Calculate the standard error for the difference in means
standard_error_diff = (sample_stddev_Competitor**2 / sample_size_Competitor + sample_stddev_Competitor**2
                       / sample_size_ABC)**0.5

# Calculate the z-score
z_score = (sample_mean_ABC - sample_mean_Competitor - hypothesized_difference) / standard_error_diff

# Calculate the p-value for a one-tailed test (greater than)
p_value = 1 - stats.norm.cdf(z_score)

# Set the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to make a decision
if p_value < alpha:
    print(f"Reject the null hypothesis. There is enough evidence to conclude 
          that ABC batteries last at least 60 minutes longer than the competitor brand (p-value = {p_value:.4f})")
else:
    print(f"Fail to reject the null hypothesis. There is not enough evidence to conclude 
          that ABC batteries last at least 60 minutes longer than the competitor brand (p-value = {p_value:.4f})")
```

### Problem_08.
A training institute would like to check the effectiveness of their training programs and if the scores
of the people trained improve post the training. A sample of 30 students was taken and their scores
were calculated before and after the training. File trainingscores.csv contains scores of the students
before and after the training. Frame and write the code to test the hypothesis of training effectiveness
for the training institute.

```python
import pandas as pd
import scipy.stats as stats

# Load the data from the CSV file
data = pd.read_csv('trainingscores.csv')

# Assuming the CSV file has two columns: 'Before' and 'After'
# You can adjust the column names accordingly

# Extract the 'Before' and 'After' scores as two arrays
scores_before = data['Before']
scores_after = data['After']

# Perform a paired t-test
t_statistic, p_value = stats.ttest_rel(scores_before, scores_after)

# Set the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to make a decision
if p_value < alpha:
    print(f"Reject the null hypothesis. There is enough evidence to conclude that the training is effective (p-value = {p_value:.4f})")
else:
    print(f"Fail to reject the null hypothesis. There is not enough evidence to conclude that the training is effective (p-value = {p_value:.4f})")
```

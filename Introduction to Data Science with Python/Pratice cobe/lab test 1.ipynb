1.write a python funcation to sum all the numbers in list

```python
def sum_list_numbers(numbers):
    total = 0

    for number in numbers:
        if isinstance(number, (int, float)):
            total += number 

    return total


my_list = [8, 2, 3, 0, 7]
result = sum_list_numbers(my_list)
print("Sum:", result)
```

2. write a python program to convert temperatures to and from celsius and fahernheit

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

celsius_temperature = 60
fahrenheit_temperature = celsius_to_fahrenheit(celsius_temperature)
print(f"{celsius_temperature}°C is {fahrenheit_temperature}°F")

fahrenheit_temperature = 45
celsius_temperature = fahrenheit_to_celsius(fahrenheit_temperature)
print(f"{fahrenheit_temperature}°F is {celsius_temperature}°C")
```

4.iv) plot the box ;identtify if there are outliers and treat them .display the plotbox before and after the outlier treatment.

```python

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=data_iris)
def treat_outliers(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    data[column] = data[column].apply(lambda x: min(upper_bound, max(lower_bound, x)))

numerical_columns = data_iris.select_dtypes(include=['float64']).columns
for column in numerical_columns:
    treat_outliers(data_iris, column)
plt.subplot(1, 2, 2)
sns.boxplot(data=data_iris)
plt.tight_layout()
plt.show()
```
